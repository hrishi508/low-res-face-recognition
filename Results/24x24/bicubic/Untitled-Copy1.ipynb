```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bicubic_tacc = []
with open('train_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        bicubic_tacc.append(float(i))
        
bicubic_vacc = []
with open('val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        bicubic_vacc.append(float(i))
        
bicubic_tloss = []
with open('train_loss.txt', 'r') as f:
    for i in list(f.read().split()):
        bicubic_tloss.append(float(i))
        
bicubic_vloss = []
with open('val_loss.txt', 'r') as f:
    for i in list(f.read().split()):
        bicubic_vloss.append(float(i))
    
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(7, 5)) ## (10, 6)
plt.subplot(1, 1, 1)
plt.grid()
plt.plot(bicubic_tacc, '--',label='Training Accuracy', linewidth=3)
plt.plot(bicubic_vacc, label='Validation Accuracy', linewidth=3)
plt.legend(loc='best', fontsize = 15) ## 20
plt.ylabel('Accuracy', fontsize = 15)  ## 20
plt.xlabel('Epochs', fontsize = 15) ## 20
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(7, 5)) ## (10, 6)
plt.subplot(1, 1, 1)
plt.ylim(top=7)
plt.grid()
plt.plot(bicubic_tloss, '--',label='Training Loss', linewidth=3)
plt.plot(bicubic_vloss, label='Validation Loss', linewidth=3)
plt.legend(loc='best', fontsize = 15) ## 20
plt.ylabel('Mean Squared Error', fontsize = 15)  ## 20
plt.xlabel('Epochs', fontsize = 15) ## 20
```
