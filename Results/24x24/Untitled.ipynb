```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
edsracc = []
with open('edsr/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        edsracc.append(float(i))
    
wdsracc = []
with open('wdsr/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        wdsracc.append(float(i))
    
srganacc = []
with open('srgan/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        srganacc.append(float(i))
    
bicubicacc = []
with open('bicubic/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        bicubicacc.append(float(i))
    
```

```python
plt.rcParams.update({'font.size': 15}) ## 20
plt.figure(figsize=(7, 5)) ## (10, 6)
plt.subplot(1, 1, 1)
plt.grid()
plt.plot(edsracc, label='EDSR', linewidth=3)
plt.plot(wdsracc, label='WDSR', linewidth=3)
plt.plot(srganacc, label='SRGAN', linewidth=3)
plt.plot(bicubicacc, label='Bicubic', linewidth=3)
plt.legend(loc='best', fontsize = 15) ## 20
plt.ylabel('Accuracy', fontsize = 15)  ## 20
plt.xlabel('Epoch', fontsize = 15) ## 20
```
