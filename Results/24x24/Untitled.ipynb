```python
import matplotlib.pyplot as plt
```

```python
edsracc = []
with open('edsr/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        edsracc.append(float(i))
    
wdsracc = []
with open('wdsr/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        wdsracc.append(float(i))
    
srganacc = []
with open('srgan/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        srganacc.append(float(i))
    
bicubicacc = []
with open('bicubic/val_acc.txt', 'r') as f:
    for i in list(f.read().split()):
        bicubicacc.append(float(i))
    
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(10, 6))
plt.subplot(1, 1, 1)
plt.plot(edsracc, label='EDSR')
plt.plot(wdsracc, label='WDSR')
plt.plot(srganacc, label='SRGAN')
plt.plot(bicubicacc, label='Bicubic')
plt.legend(loc='best', fontsize = 20)
plt.ylabel('Accuracy', fontsize = 20)
plt.xlabel('epoch', fontsize = 20)
```
