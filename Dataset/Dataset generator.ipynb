```python
#MTCNN code that bypasses cases where mtcnn does not detect faces

import os
import shutil
import mtcnn
from mtcnn.mtcnn import MTCNN
import cv2
src = 'Sample Dataset/' #path to the dataset
dest1 = 'final_gen_data_10/train/' #10 per subj, path to the folder where you want to store the generated training images
dest2 = 'final_gen_data_10/val/' #3 per subj, path to the folder where you want to store the generated validation images
dest3 = 'final_gen_data_10/test/' #2 per subj, path to the folder where you want to store the generated test images
i = 0
x = 0
y = 0
z = 0
detector = MTCNN()
flag = 0

def process(pixels):
    global flag
    results = detector.detect_faces(pixels)
    if len(results) == 0:
        flag = 1
        return 0
    x1, y1, width, height = results[0]['box']
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height
    img = pixels[y1:y2, x1:x2]
    img = cv2.resize(img, (160,160), interpolation=cv2.INTER_CUBIC)
    return img

for folder in os.listdir(src):
    folder_path = src + folder + '/'
    i = 0
    print(folder)
    for file in os.listdir(folder_path):
        file_path = folder_path + file
        img = cv2.imread(file_path)
        img = process(img)
        if(flag == 1):
            flag = 0
            continue
        
        if i<10:
            destination = dest1 + str(x) + '.jpg'
            cv2.imwrite(destination, img)
            x += 1
            
        elif i<13:
            destination = dest2 + str(y) + '.jpg'
            cv2.imwrite(destination, img)
            y += 1
            
        elif i<15:
            destination = dest3 + str(z) + '.jpg'
            cv2.imwrite(destination, img)
            z += 1
            
        else:
            break
            
        i += 1
```
